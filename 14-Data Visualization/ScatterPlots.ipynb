# Scatter plots

To create a simple scatter plot, we use the `sns.scatterplot` command and specifiy the value for:

* the horizontal x-axis (`x=insurance_data['bmi']`), and

* the vertical y-axis(`y=insurance_data["charges"]`)

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

## Color-coded scatter plots

We can use scatter plots to display the relationships between (not two, but...) Three variables! one way of doing this by color coding the points

For instance, to understand how smoking affects the relationship between BMI and insurance costs, we can color-code the points by `smoker`, and plot the other two columns(`bmi`, `charges`)on the axis

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

The scatter plot shows that while nonsomkers to tend to pay slightly more with increasing BMI, smokers pay much more.

To futher emphasize this fact, we can use the `sns.lmplot` command to add two regression lines, corresponding to smokers and nonsmokers.

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
```

The `sns.lmplot` command works slightly differently than the commands above that. 

* Instead of setting `x=insurance_data['bmi']` to select the `bmi` column in  `insurance_data`, we set `x="bmi"` to speciy the name of the column only. 

* Similary, `y="charges"` and `hue="smoker"` also contain the names of columns

* we specify the dataset with `data=insurance_data`

```python
sns.swarmplot(x = insurance_data['smoker'],
              y=insurance_data['charges'])
```
