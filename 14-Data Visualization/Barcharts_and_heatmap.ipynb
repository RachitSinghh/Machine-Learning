```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
flight_filepath = "data/flight_delays.csv"
flight_data = pd.read_csv(flight_filepath, index_col="Month")
```

```python
flight_data
```

# Bar Chart

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title 
plt.title("Average arrival delay for spirit airlines flights, by month")

# Bar chart showing average arrival delay for spirit airlines flight by month
sns.barplot(x=flight_data.index, y=flight_data['NK'])

# add label for vertical axis 
plt.ylabel("Arrival delay (in minutes)")
```

```python
# Bar Chart showing average arrival delay for spirit airlines flights by month
sns.barplot(x = flight_data.index, y=flight_data['NK'])
```

* `sns.barplot` - this tells the notebook that we want to creaet a bar chart\

* `x=flight_data.index` - this determines what to use on the horizontal axis. In this case, we have selected the column that indexes the rows (in thi case,, the column containing the months).

* `y=flight_data['NK']` - This sets the column in the data that will used to determine the height of each bar. In this case, we select `NK` column

# Heatmap

```python
plt.figure(figsize=(14,7))

plt.title("Average arrival delay for each airline, by month")

# heatmap showing average arrival delay for each airline

sns.heatmap(data = flight_data, annot=True)

# add lable 
plt.xlabel("Airline")
```

```python
# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=flight_data, annot=True)
```

* `sns.heatmap` - This tells the notebook that we want to create a heatmap. 

* `data=flight_data` - This tells the notebook to use all of the entries in `flight_data` to create the heatmap.

* `annot=True` This ensures that the values for each cell appear on the chart. (Leaving this out removes the numbers from each of the cells!)
