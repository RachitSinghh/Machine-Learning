# Abstraction

Abstraction is the concept of hiding the complex implementation details and showing only the necessary feature of an object. This helps in reducing programming complextiy and effort

```python
from abc import ABC, abstractmethod

## Abstract class

class Vechicle(ABC):
    def drive(self):
        print("The vehicle is used for driving")
    
    @abstractmethod # The abstract method is like giving the necessary feature of an object and it is just hiding the complex implementation details.
    def start_engine(self):
        pass

class Car(Vechicle):
    def start_engine(self):
        print("Car engine started")

def operate_vehicle(vechicle):
    vechicle.start_engine()
    vechicle.drive()

car = Car()
operate_vehicle(car)
```
